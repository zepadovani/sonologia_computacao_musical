```python
sr = 44100                # sample rate
dur = 2                   # duracao em segundos
t = np.arange(0,dur*sr)   # serie temporal (em samples)
freq = 440                # frequencia
per = 1/freq              # periodo
fase = 0                  # fase
A = 0.5                   # amplitude
```

$y = A \cdot sin(2 \pi f t + ϕ)$

```python
y = A * np.sin(2 * np.pi * freq * t/sr + fase)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(y, sr=sr) #plor de todo o áudio gerado
```

```python
plt.figure(figsize=(14, 5))
plt.plot(t,y) #plot de todo o áudio gerado (via librosa)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(y[0:int((1/freq)*sr)], sr=sr) #plot de todo o áudio gerado (via librosa)
```

```python
ipd.Audio(y, rate=sr)
```

```python
last_ind = int(per*sr)

plt.figure(figsize=(14, 5))
plt.plot(t[:last_ind],y[:last_ind]) #plot de todo o áudio gerado
```

```python
sr = 44100                # sample rate
dur = 10                   # duracao em segundos
t = np.arange(0,dur*sr)   # serie temporal (em samples)

fr1 = 440                   # frequencia
per1 = 1/freq               # periodo
fa1 = 0                     # fase
A1 = 0.5                    # amplitude
y1 = A1 * np.sin(2 * np.pi * fr1 * t/sr + fa1)

fr2 = 441                   # frequencia
p2 = 1/freq                 # periodo
fa2 = 0                     # fase
A2 = 0.5                    # amplitude
y2 = A2 * np.sin(2 * np.pi * fr2 * t/sr + fa2)
```

```python
lastsamp=12000
plt.figure(figsize=(14, 5))
plt.plot(t[:lastsamp],y1[:lastsamp])
plt.plot(t[:lastsamp],y2[:lastsamp]) #plot de todo o áudio gerado (via librosa)
plt.plot(t[:lastsamp],(y1+y2)[:lastsamp])
```
